# Q5: Explain ImportError. What is ModuleNotFoundError?
"""
# Explanation of ImportError and ModuleNotFoundError
"""
ImportError is raised when an imported module cannot be found or loaded due to various reasons such as incorrect module name, missing module file, or failure to initialize the module correctly.

ModuleNotFoundError is a subclass of ImportError introduced in Python 3.6 specifically to indicate that a module could not be found. It provides a more specific error message when the module specified in an import statement cannot be located.

Example:
try:
    import non_existing_module
except ImportError as e:
    print(f"ImportError occurred: {e}")
"""

# Q6: List down some best practices for exception handling in Python.
"""
# Best practices for exception handling in Python
- Use specific exception handling: Catch specific exceptions rather than broad exceptions like Exception.
- Avoid empty except blocks: Handle exceptions explicitly and avoid catching all exceptions unless necessary.
- Clean up using finally: Use finally blocks for cleanup tasks that should execute regardless of whether an exception occurred.
- Raise exceptions appropriately: Raise exceptions with informative error messages to aid debugging and error resolution.
- Handle exceptions close to the point of occurrence: Catch exceptions as close to where they occur as possible to maintain code clarity and reduce unexpected behavior.
