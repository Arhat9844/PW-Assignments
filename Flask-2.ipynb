# Q1: Explain GET and POST methods.

"""
ANSWER:
GET and POST are HTTP methods used to request data from a server and to send data to a server, respectively.

- GET: 
  - Used to request data from a specified resource.
  - The data is sent as part of the URL.
  - It is a read-only operation and does not affect the data on the server.
  - Suitable for fetching data.

- POST:
  - Used to send data to a server to create/update a resource.
  - The data is sent in the body of the HTTP request.
  - It can modify the data on the server.
  - Suitable for sending data such as form submissions.
"""

# Q2: Why is request used in Flask?

"""
ANSWER:
The request object in Flask is used to handle the incoming HTTP requests. It allows you to access the data sent by the client, such as form data, query parameters, headers, and cookies. It is essential for processing and responding to client requests in a web application.
"""

# Q3: Why is redirect() used in Flask?

"""
ANSWER:
The redirect() function in Flask is used to redirect the client to a different URL. This is useful for directing users to another page after an action has been performed, such as after submitting a form or logging in. It helps in maintaining a smooth user experience and managing URL navigation.
"""

# Q4: What are templates in Flask? Why is the render_template() function used?

"""
ANSWER:
Templates in Flask are files (usually HTML) that define the structure of web pages. They allow you to separate the presentation logic from the application logic, making the code cleaner and more maintainable. The render_template() function is used to render these templates and pass data from the Flask application to the template for dynamic content generation.
"""

# Q5: Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

from flask import Flask, request, jsonify
from IPython.display import display, Image

app = Flask(__name__)

@app.route('/api', methods=['GET', 'POST'])
def api():
    if request.method == 'GET':
        return jsonify({"message": "This is a GET request"})
    elif request.method == 'POST':
        data = request.json
        return jsonify({"message": "This is a POST request", "data": data})

if __name__ == '__main__':
    app.run(debug=True)
