Part 1: Theoretical Questions
Q1: What are the different data structures used in TensorFlow? Give some examples.
Answer:

TensorFlow uses several data structures, with the primary ones being:

Tensors: Multidimensional arrays that represent the basic unit of data in TensorFlow. Examples include scalar (0-D tensor), vector (1-D tensor), matrix (2-D tensor), and higher-dimensional tensors.
Variables: Mutable tensors that store parameters that can be updated during training.
Constants: Immutable tensors that hold fixed values.
Datasets: Used for input pipelines, providing an efficient way to handle large datasets.

Q2: How does the TensorFlow constant differ from a TensorFlow variable? Explain with an example.
Answer:

TensorFlow Constants: Immutable tensors whose values cannot be changed after they are created. They are often used for storing fixed values.
TensorFlow Variables: Mutable tensors whose values can be updated. They are used to store model parameters that need to be updated during training.
Example:

import tensorflow as tf

# TensorFlow constant
const = tf.constant([1, 2, 3])
print("Constant:", const)

# TensorFlow variable
var = tf.Variable([1, 2, 3])
print("Variable before update:", var)

# Updating the variable
var.assign([4, 5, 6])
print("Variable after update:", var)

Q3: Describe the process of matrix addition, multiplication, and element-wise operations in TensorFlow.
Answer:

Matrix Addition: Adding corresponding elements of two matrices.
Matrix Multiplication: Performing a dot product between rows of the first matrix and columns of the second matrix.
Element-wise Operations: Performing operations on each corresponding element of two matrices, such as element-wise multiplication or division.
Example:

import tensorflow as tf

# Creating two matrices
matrix1 = tf.constant([[1, 2], [3, 4]])
matrix2 = tf.constant([[5, 6], [7, 8]])

# Matrix addition
add_result = tf.add(matrix1, matrix2)

# Matrix multiplication
mul_result = tf.matmul(matrix1, matrix2)

# Element-wise multiplication
elem_mul_result = tf.multiply(matrix1, matrix2)

print("Matrix Addition:\n", add_result)
print("Matrix Multiplication:\n", mul_result)
print("Element-wise Multiplication:\n", elem_mul_result)
