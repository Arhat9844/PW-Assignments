"""
ANSWER:
Flask is a lightweight WSGI web application framework in Python. It is designed with simplicity and flexibility in mind, making it a popular choice for small to medium-sized applications, as well as for prototyping and experimentation.

Advantages of Flask Framework:
1. Lightweight and Minimalistic: Flask has a simple core with easily extendable functionality through add-ons.
2. Modular and Flexible: You can use Flask components independently or together.
3. Built-in Development Server: Flask includes a built-in server for easy testing and development.

5. Support for Unit Testing: Flask has integrated support for unit testing.

# Q2: Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

from flask import Flask
from IPython.display import display, Image

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(debug=True)

# Run the Flask app and take a screenshot of the output.
# ANSWER:
display(Image(filename='path_to_hello_world_screenshot.png'))

# Q3: What is App routing in Flask? Why do we use app routes?

"""
ANSWER:
App routing in Flask is the mechanism to map URLs to functions in your web application. The @app.route decorator is used to bind a URL to a function.

Why do we use app routes?
1. URL Mapping: Routes map URLs to specific functions that handle the requests.
2. Simplifies Navigation: It allows for easy navigation and clean URLs.
3. Dynamic Content: Routes can be parameterized to handle dynamic content and inputs.
"""

# Q4: Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
# - Company Name: ABC Corporation
# - Location: India

# Q5: What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

"""
ANSWER:
The url_for() function is used in Flask for URL building. It is used to build a URL to a specific function dynamically.
"""

@app.route('/contact')
def contact():
    return 'Contact us at 999-999-9999'

@app.route('/demo')
def demo():
    home_url = url_for('hello_world')
    welcome_url = url_for('welcome')
    contact_url = url_for('contact')
    return f'''
    Home URL: {home_url}<br>
    Welcome URL: {welcome_url}<br>
    Contact URL: {contact_url}
    '''

if __name__ == '__main__':
    app.run(debug=True)
